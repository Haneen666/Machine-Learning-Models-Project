```python
import tensorflow as tf
```

```python
(x_train,y_train),(x_test,y_test) =tf.keras.datasets.mnist.load_data()
print(x_train[1])
```

```python
x_train=x_train/255.0
x_test=x_test/255.0
print (x_train[1])
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(128,activation='relu'),
    tf.keras.layers.Dense(10,activation='softmax')
])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,
          epochs=1,
          batch_size=64)
```

```python
test_loss, test_acc = model.evaluate(x_test,y_test)
print("دقة النموذج على بيانات الاختبار:",test_acc)
print("قيمة خسارة النموذج على البيانات الاختبار", test_loss)
```

```python
predictins = model.predict(x_test[:5])
for i, pred in enumerate(predictins):
  print(f"الصورة{i},نموذج التوقع: {pred.argmax()},القيمة الحقيقية:{y_test[i]}")
```
